```python
from numpy import *
import numpy as np
import scipy
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

```

```python
sequences = pd.read_csv('truck_sequences.csv')
del sequences['Unnamed: 0']
sequences = sequences.transpose()
```

```python
orders = pd.read_csv('truck_orders.csv')
orders = orders.transpose()
```

Here are 1000 orders where each order is a random number between 0 and 5 for each truck type

```python
orders.head(5)
```

And here are randomly generated sequences for each order. There are a good amount of NaNs right now since the sequences vary a lot in length.

```python
sequences.head(5)
```

Let's concatenate the orders & sequences, which will be our full set of training data.

```python
training_data = pd.concat([orders,sequences],axis=1)
training_data.columns = np.arange(118)
```

We will consider the first 100 sequences as 'good'.

```python
labels = pd.DataFrame(np.zeros(len(sequences)))
labels.loc[0:99] = 1
labels.columns = ['label']

training_data['labels'] = labels.values
```

We now have a sample of 1000 order - sequence combinations, with 100 labeled as 'good'

```python
training_data.head()
```

^ The above represents our full training data of 1,000 order-sequence samples. 100 are labeled as 'good' being 1. 

The first 26 columns represent the order profile - the quantities of each truck model (A-Z).
The next 92 columns represent all possible sequence values. This is 92 columns because that was the length of the longest sequence (largest order) randomly generated. The final column is our label of whether the order-sequence combo was 'good' or not.

The next steps here would probably be dependent on the model we want to try. Computing n-gram metrics would be useful for a SVM classifier. Establishing a distance metric would be useful for a KNN classifier. Using PCA would be useful to esablish a smaller finite set of features to use in a Random Forest classifier. All of these have the goal of classifying (with probabilities of being 'good') all possible sequence-order combinations for a new order.
